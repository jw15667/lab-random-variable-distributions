```python
sns.displot(data['Monthly Premium Auto'])
plt.show()
```

###  log trans

```python
def log_transform(x):
    if np.isfinite(x) and x!=0:
        return np.log(x)
    else:
        return np.NAN
```

```python
data['Monthly_Premium_Auto_box'] = data['Monthly Premium Auto'].apply(log_transform)
```

```python
sns.displot(data['Monthly_Premium_Auto_box'])
plt.show()
```
